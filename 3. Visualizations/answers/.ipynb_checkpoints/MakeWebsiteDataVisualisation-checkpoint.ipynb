## Make geojson profiles into correct format for shoreline monitor

```python
import json
import geojson
import pathlib
import shapely.geometry
import shapely.affinity
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
%matplotlib inline
```

```python
N = matplotlib.colors.Normalize(-5, 5, clip=True) # clip between -5 and 5 m/y
```

```python
# open file
folder = r'P:\11200481-sandy-solution-bacton\02_Smoothing_Global_Coastlines\Version 16102018\Changerates'
savefolder = r'P:\11200481-sandy-solution-bacton\02_Smoothing_Global_Coastlines\Version 16102018\ShorelineMonitor points'
fnames = os.listdir(folder)

pointfeatures = []
for fname in np.sort(fnames):
    print fname
    with open(os.path.join(folder, fname)) as f:
        data = json.load(f)

    for feature in data['features']:
        if feature['geometry'] == None:
            del feature
        else:
            if feature['properties']['change_rate'] == -999:
                del feature
            else:
                # read the geometry and store it in a property
                feature['properties']['line'] = feature['geometry']
                # this does not seem to match the centroid...
                feature['properties']['transect_origin'], feature['geometry']
                # convert to shapely
                line = shapely.geometry.asShape(feature['geometry'])

                if len(line.xy[0]) != 2:
                    del feature
                else:
                    # get point halfway transect
                    p = shapely.geometry.mapping(line.interpolate(0.5*line.length))

                    # convert coordinates to array
                    coords = np.array(line.coords)
                    # compute the dx and dy of the line
                    dx, dy = np.squeeze(np.diff(coords, axis=0))
                    # compute the center (not used I think..)
                    #cx, cy = np.squeeze(np.mean(coords, axis=0))
                    # compute the angle
                    theta = np.arctan2(dy, dx)
                    transectID = feature['properties']['transect_id']
                    feature['properties']['theta'] = theta
                    change_rate = feature['properties']['change_rate']
                    # scale to -1, 1
                    normalized_change_rate = N(change_rate) * 2 - 1.0
                    
                    if change_rate == -999:
                        normalized_change_rate = 0
#                     feature['properties']['normalized_change_rate'] = normalized_change_rate

#                     # move along center (shift outwards) (hope that the line has a direction)
#                     shifted_line = shapely.affinity.translate(
#                         line, 
#                         xoff=dx/2., 
#                         yoff=dy/2., 
#                         zoff=0.0
#                     )
#                     scaled_line = shapely.affinity.scale(
#                         shifted_line,
#                         xfact=normalized_change_rate,
#                         yfact=normalized_change_rate,
#                         origin=line.centroid
#                     )
#                     feature['geometry'] = scaled_line.__geo_interface__

                    p = geojson.Feature(geometry=p, properties={'Transect_id':transectID,'Angle':theta,'Value':change_rate,'NormalizedValue':normalized_change_rate})
                    pointfeatures.append(p)
            
# store point features
with open(os.path.join(savefolder, 'TransectsAsPoints.geojson'), 'w') as f:
    json.dump(geojson.FeatureCollection(pointfeatures), f)
```

```python
# normalize distribution, long tailed
rates = [feature['properties']['change_rate'] for feature in data['features']]
p2_5 = np.percentile(rates, 10)
_ = plt.hist(rates, range=(p2_5, -p2_5), bins=80)
```

```python
# make timeseries json
folder = r'P:\11200481-sandy-solution-bacton\02_Smoothing_Global_Coastlines\Version 16102018\Changerates'
savefolder = r'P:\11200481-sandy-solution-bacton\02_Smoothing_Global_Coastlines\Version 16102018\ShorelineMonitor timeseries'

fnames = os.listdir(folder)
fts = []
for fname in np.sort(fnames)[80:]:
    print fname
    with open(os.path.join(folder, fname)) as f:
        data = json.load(f)

    for feature in data['features']:
        if feature['geometry'] == None:
            del feature
        else:
            if feature['properties']['change_rate'] == -999:
                del feature
            else:
                ft = {}
                ft['distances'] = feature['properties']['distances']
                ft['dt'] = feature['properties']['dt']
                ft['intercept'] = feature['properties']['dt']
                ft['b_unc'] = feature['properties']['b_unc']
                ft['contintent'] = feature['properties']['continent']
                ft['country_name'] = feature['properties']['country_name']
                ft['changerate'] = feature['properties']['change_rate']
                ft['change_rate_unc'] = feature['properties']['change_rate_unc']
                ft['outliers_1'] = feature['properties']['outliers_1']
                ft['outliers_2'] = feature['properties']['outliers_2']
                ft['transectID'] = feature['properties']['transect_id']
                ft['flag_sandy'] = feature['properties']['flag_sandy']
                fts.append(ft)
                
    # store
    with open(os.path.join(savefolder, fname.replace('geo','')), 'w') as f:
        json.dump(fts, f)
```
